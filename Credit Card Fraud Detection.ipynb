```python
#Import the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Load the dataset
credit_data=pd.read_csv("C:\\Users\\kotha\\OneDrive\\Desktop\\creditcard.csv")
```

```python
credit_data['Class'].value_counts()
```

```python
#The dataset is highly inbalenced
#0->Normal transaction
#1->fraud transaction
```

```python
#Seperate the data for analysis
legit=credit_data[credit_data.Class==0]
fraud=credit_data[credit_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_data.groupby('Class').mean()
```

```python
#dealing with the unbalanced data
#undersampling
```

```python
#Build a sample dataset containing similar distribution of normal and fraud transaction
legit_sample=legit.sample(n=492)
```

```python
#concatenating two dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting the data in to feauters and  target
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#split the training data and testing data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#MODEL TRAINING
#LOGISTIC REGRRESSION
```

```python
model=LogisticRegression()
```

```python
print(model)
```

```python
#training logistic regression model with training data
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
#Accuracy score
```

```python
#accuracy score on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
    print("Accuracy on training data : ",training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on test data : ",test_data_accuracy)
```
